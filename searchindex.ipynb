```python

from urllib.request import urlopen
from contextlib import closing

URL = 'ftp://ftp.nasdaqtrader.com/SymbolDirectory/nasdaqlisted.txt'

symbols = {}
with closing(urlopen(URL)) as raw:
    for line in raw:
        s = line.decode().split('|')
        if s[3] == 'N' and s[4] == 'N' and s[6] == 'N':
            symbols[s[0]] = s[1].split(' - ')[0]

URL = 'ftp://ftp.nasdaqtrader.com/SymbolDirectory/otherlisted.txt'

with closing(urlopen(URL)) as raw:
    for line in raw:
        s = line.decode().split('|')
        if s[0] in symbols:
            print("duplicate:", s[0])
        if s[4] == 'N' and s[6] == 'N':
            symbols[s[0]] = s[1].split(' - ')[0]

import pickle
pickle.dump(symbols, open("symbols.pickle", "wb"))
```
